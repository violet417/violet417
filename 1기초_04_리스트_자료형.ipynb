```python
# 데이터들이 연속적으로 저장되는 형태의 자료형
# 리스트 자료형
a = []
b = [1,3,5,4,2]
c = ['text1','text2']
d = [1,3,5,'text1','text2']
e = [1,2,3,[4,5]]
```

```python
# 리스트 인덱싱
a = [1,3,5,4,2]
print(a[0])
print(a[-1])
```

```python
# 리스트 슬라이싱
a = [1,2,3,4,5]
print(a[0:2])
print(a[2:])
```

```python
# 리스트 덧셈 (뺄셈은 불가능하다.)
a = [1,2,3]
b = [4,5,6]
print(a + b)
```

```python
# 리스트 곱셈
a = [1,2,3]
print(a * 3)
```

```python
# 리스트 길이 구하기
a = [1,2,3]
print(len(a))
```

```python
# 리스트 값 변경하기
a = [1,3,5,4,2]
a[1] = 10
print(a)
```

```python
# 리스트 값 삭제하기
a = [1,3,5,4,2]
del a[1]
print(a)
```

```python
a = [1,3,5,4,2]
del a[2:]
print(a)
```

```python
# list 내장함수
a = "Pyhon"
list(a)
```

```python
# 리스트 전용함수: append (맨 끝에 값 추가)
a = [1,3,5,4,2]
a.append(7)
print(a)
```

```python
# 리스트 전용함수: sort (오름차순 정렬)
a = [1,3,5,4,2]
a.sort()
print(a)
```

```python
a = ['car', 'apple', 'banana']
a.sort()
print(a)
```

```python
# 리스트 전용함수: sort (내림차순 정렬)
a = [1,3,5,4,2]
a.sort(reverse=True)
print(a)
```

```python
a = ['car', 'apple', 'banana']
a[::-1]
print(a)
```

```python
# 리스트 전용함수: reverse (리스트를 거꾸로 출력)
a = ['car', 'apple', 'banana']
a.reverse()
print(a)
```

```python
# 리스트 전용함수: index (없는 값을 찾는 경우 오류 발생)
a = ['car', 'apple', 'banana']
print(a.index('apple'))
```

```python
# 리스트 전용함수: insert
a = [1,3,5,4,2]
a.insert(1,7)    # 1번째에 7 삽입
print(a)
```

```python
# 리스트 전용함수: remove
a = [1,3,5,4,2]
a.remove(5)    # 삭제할 값 
print(a)
```

```python
# 리스트 전용함수: pop (가장 뒤에 있는 값을 추출)
list1 = [1,3,5,4,2]
a = list1.pop()
print(list1)
print(a)
```

```python
list1 = [1,3,5,4,2]
a = list1.pop(0)    # 0번째 값 추출
print(list1)
print(a)
```

```python
# 리스트 전용함수: count
a = [1,3,5,4,2,2,2]
print(a.count(2))
```

```python
# 리스트 전용함수: extend
a = [1,3,5,4,2]
a.extend([1,1,1])    # a = a + [1,1,1]와 같음
print(a)
```
