```python
# 문자열 덧셈 (뺄셈은 불가능하다.)
a = 'Python'
b = ' is too fun!'
print(a + b)
```

```python
# 문자열 곱셈
a = 'Python'
print(a * 3)
```

```python
# 문자열 길이 구하기
a = 'Python'
print(len(a))
```

```python
# 문자열 인덱싱
a = 'Python'
print(a[0])
print(a[-1])
```

```python
# 문자열 슬라이싱
a = 'Python is too fun!'
print(a[2:9])
print(a[:9])
print(a[3:])
```

```python
# 문자열 전용함수: format
text = "I eat {} apples".format(3)
print(text)
```

```python
text = "I eat {} apples and {} oranges".format(3, 5)
print(text)
```

```python
text = "I eat {0} apples and {0} oranges".format(3, 5)
print(text)
```

```python
text = "{:>10}".format("Hi")
print(text)
```

```python
text = "{:<10}".format("Hi")
print(text)
```

```python
text = "{:^10}".format("Hi")
print(text)
```

```python
text = "{:=^10}".format("Hi")
print(text)
```

```python
text = "{:.2f}".format(3.141592)
print(text)
```

```python
# 문자열 전용함수: count
text = "Python:is:too:fun!"
print(text.count('o'))
```

```python
# 문자열 전용함수: find (없는 값을 찾을 경우 -1로 표시)
text = "Python is too fun!"
print(text.find('o'))
print(text.find('z'))
```

```python
# 문자열 전용함수: join
text = ",".join("abcde")
print(text)
```

```python
# 문자열 전용함수: upper
text = "Python is too fun!"
print(text.upper())
```

```python
# 문자열 전용함수: lower
text = "Python is too fun!"
print(text.lower())
```

```python
# 문자열 전용함수: lstrip
text = "    Hello   "
print(text.lstrip())
```

```python
# 문자열 전용함수: rstrip
text = "    Hello   "
print(text.rstrip())
```

```python
# 문자열 전용함수: strip
text = "    Hello   "
print(text.strip())
```

```python
# 문자열 전용함수: replace
text = "Python is too fun!"
print(text.replace("Python", "Java"))
```

```python
# 문자열 전용함수: split
text = "Python is too fun!"
print(text.split())
```

```python
text = "Python:is:too:fun!"
print(text.split(':'))
```

```python
# 문자열 전용함수: startswith
text = "Python is too fun!"
text.startswith("Python")
```

```python
# 문자열 전용함수: endswith
text = "Python is too fun!"
text.endswith("fun!")
```

```python
# 문자열 전용함수: isalpha
text = 'Hello'
print(text.isalpha())
```

```python
text = '안녕하세요'
print(text.isalpha())
```

```python
text = 'Hello Wolrd!'
print(text.isalpha())
```

```python
text = 'Hello300'
print(text.isalpha())
```

```python
# 문자열 전용함수: isdigit
text = '300'
print(text.isdigit())
```

```python
# 문자열 전용함수: isalnum
text = 'hello300'
print(text.isalnum())
```

```python
# 문자열 전용함수: isspace
text = '    '
print(text.isspace())
```
